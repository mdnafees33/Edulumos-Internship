```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df = pd.read_csv("train.csv")
```

```python
print("First 5 rows of the dataset:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
sns.countplot(x='Survived', data=df, palette='coolwarm')
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='coolwarm')
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='coolwarm')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df, palette='coolwarm')
plt.title('Age vs Survival')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.barplot(x='Embarked', y='Survived', data=df, palette='coolwarm')
plt.title('Survival Rate by Embarked Port')
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
sns.barplot(x='FamilySize', y='Survived', data=df, palette='coolwarm')
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
print("\nKey Insights:")
print("- Females had a higher survival rate than males.")
print("- Passengers in 1st class were more likely to survive.")
print("- Younger passengers had a slightly higher chance of survival.")
print("- People who embarked from 'C' port survived more on average.")
print("- Smaller families (1–3 members) had better survival rates.")
```
