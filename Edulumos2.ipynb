```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df = pd.read_csv("/content/mymoviedb.csv", engine='python', on_bad_lines='skip')
```

```python
print("üìÑ First 5 Rows:")
print(df.head())
```

```python
print("\nüìä Dataset Info:")
df.info()
```

```python
print("\nüîç Missing Values:")
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.columns)
```

```python
print("\nüìà Basic Statistics:")
print(df.describe(include='all'))
```

```python
df['release_year'] = pd.to_numeric(df['Release_Date'].str.split('-').str[0], errors='coerce')
plt.figure(figsize=(25, 6))
sns.countplot(data=df, x='release_year', palette='pastel', order=df['release_year'].value_counts().sort_index().index)
plt.title("Content Added by Release Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')
plt.figure(figsize=(6, 6))
sns.histplot(data=df, x='Vote_Average', bins=20, kde=True, color='skyblue')
plt.title("Distribution of Vote Average")
plt.xlabel("Vote Average")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Genre'] = df['Genre'].fillna("")
genres = df['Genre'].str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(6, 6))
sns.barplot(x=genres.values, y=genres.index, palette='viridis')
plt.title("Top 10 Popular Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
print("\nüìä Summary Insights:")
print(f"- Total Titles: {len(df)}")
print(f"- Number of Countries: N/A (Country column not available)")
print(f"- Most Common Rating (Vote Average): {df['Vote_Average'].mode()[0]}")
print(f"- Most Common Genre: {genres.index[0]}")

print("\n‚úÖ Data analysis and visualization complete!")
```
